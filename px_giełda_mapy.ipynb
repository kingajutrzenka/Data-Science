# Notowania gie≈Çdowe

```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
def fetch_financial_data(company='AMZN'):
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')

df = fetch_financial_data('AMZN')
df.head()
```

```python
df = df.reset_index()
#ustawianie index√≥w 0, 1...

px.area(data_frame=df, x='Date', y='Close', log_y=True, title="wykres sp√≥≈Çki amazon")
```

# Flights data

```python
from seaborn import load_dataset
df = load_dataset('flights')
df.head()
```

```python
px.bar(data_frame = df, y='year', x='passengers', title="Wykres pasa≈ºer√≥w", color='year', orientation='h')
 #orientation - h wykres poziomy trzeba te≈º zamieniƒá x i y
```

```python
px.histogram(data_frame=df, x='year', y='passengers', color='year', nbins=50)
#nbins - liczba s≈Çupk√≥w
```
