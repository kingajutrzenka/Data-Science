**Matplotlib**

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
fig = plt.figure()
fig.show()
```

```python
plt.plot([0, 1], [0, 3])
# 0,1 wartości na osi X, 0,3 wartości na osi Y
# jak chcemy bez tego opisu to musimy przypisać do zmiennje
```

```python
_ = plt.plot([0, 1], [0, 3], label ='line', color= 'green') # label - legenda
_ = plt.legend() # wywołanie legendy
plt.xlabel('x') # nazwa osi x
plt.ylabel('y') # nazwa osi y
plt.title('title')
_ = plt.show()
```

```python
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)
plt.plot(x, x_sin, label = 'sin')
plt.plot(x, x_cos, label = 'cos')
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(12, 6))
plt.subplot(131) # 1 - liczba wierszy 2 - liczba kolumn 1 - wykres pierwszy
plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(132) # wykres drugi
plt.plot([0, 1, 2], [1, 2, 3], 'r--')
plt.subplot(133)
plt.plot([0, 1, 2], [1, 2, 3], 'go')
plt.show()
```

```python
plt.bar(x=[0, 1, 2, 3], height=[1, 2, 3, 4])
plt.show()
```

```python
plt.bar(x=['a', 'b', 'c', 'd'], height=[1, 2, 3, 4], color = ['r', 'y', 'g', 'b' ])
plt.show()
```

```python
x1 = np.random.randn(100)
x2 = np.random.randn(100)
plt.scatter(x1, x2)
plt.show()
```

```python
x3 = np.random.randn(100) * 50
plt.scatter(x1, x2, s = x3)
plt.show()
```

```python
plt.scatter(x1, x2, s = x3, alpha = 0.5)
plt.show()
```

```python
data = np.random.rand(10, 10)
sns.heatmap(data, annot=True)
plt.show()
```
