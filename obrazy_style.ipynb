```python
fig = plt.figure(figsize = (12, 10))
plt.subplot(221)
plt.imshow(mpimg.imread("IMG_7591.jpeg"))
plt.axis('off')
plt.subplot(222)
plt.imshow(mpimg.imread("IMG_7592.png"))
plt.axis('off')
plt.subplot(223)
plt.imshow(mpimg.imread("IMG_7593.jpeg"))
plt.axis('off')
plt.subplot(224)
plt.imshow(mpimg.imread("IMG_7594.png"))
plt.axis('off')
```

```python
fig = plt.figure(figsize = (12, 10))
for idx, img in enumerate (['IMG_7591.jpeg', 'IMG_7592.png', 'IMG_7593.jpeg', 'IMG_7594.png']):
    plt.subplot(140 + idx + 1)
    plt.imshow(mpimg.imread(img))
    plt.axis('off')
```
