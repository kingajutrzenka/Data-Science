# Wprowadzenie

plotly =/= plotly express

```python
import plotly
import plotly.graph_objects as go
```

```python
# wykres slupkowy
fig = go.Figure(
    data=go.Bar(y=[2, 3, 1, 4])
)
fig.show()
```

```python
fig = go.Figure(
    data=go.Bar(y=[2, 3, 1, 4]),
    layout=go.Layout(title={'text': 'Wykres słupkowy'})
)
fig.write_html('bar.html')
fig.show()
```

# dane diamenty

```python
from seaborn import load_dataset
df = load_dataset('diamonds')
df.head()
```

```python
dff = df.cut.value_counts()
dff = dff.reset_index()
dff.columns = ['type', 'count']
dff
```

```python
fig = go.Figure(
    data = go.Pie(labels=dff['type'], values=dff['count'], hole=0.5),
    layout=go.Layout(title={'text': 'Wykres słupkowy'}, width=800, height=600)
)
fig.show()
#labels to typy (w zasadzie to etykieta, ale ciii)
fig.write_html('pie.html')
```

# Diagram Sankey'a

```python
#służy np. do sprawdzenia przepływu klientów
```

```python
data = [go.Sankey(node = dict(label=['Nonchurn_2018', 'Churn_2018', 'Nonchurn_2019', 'Churn_2019']), # Indexy 0, 1, 2, 3
                  link=dict(source=[0, 0, 1, 1], #skąd idą
                            target=[2, 3, 2, 3], # dokąd idą
                            value=[65, 12, 18, 5]))]
""" Z nonchurn_2018 65 osób zostało aktywnych -> nonchurn_2019
    Z nonchurn_2018 12 osób odeszło -> churn_2019
    Z churn_2018 18 osób wróciło -> nonchurn_2019
    Z churn_2018 5 osób nadal nie wróciło -> churn_2019"""

fig = go.Figure(data=data)
fig.show()
```

```python
data = [go.Sankey(node = dict(label=['Nonchurn_2018', 'Churn_2018', 'Nonchurn_2019', 'Churn_2019', 'Nonchurn_2020', 'Churn_2020']), # Indexy 0, 1, 2, 3, 4, 5
                  link=dict(source=[0, 0, 1, 1, 2, 3], #skąd idą
                            target=[2, 3, 2, 3, 4, 5], # dokąd idą
                            value=[65, 12, 18, 5, 83, 17]))]

fig = go.Figure(data=data)
fig.show()
```

# Wykres świecowy

```python
def fetch_financial_data(company='AMZN'):
    """Pobiera dane dotyczące notowań spółek na giełdzie według symboli na serwisie stooq.pl"""
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')

df = fetch_financial_data()
df = df.reset_index()
df = df[df.Date > '2025-07-01']
df.head()
```

```python
fig = go.Figure(data=go.Candlestick(x=df.Date, open=df.Open, high=df.High, low=df.Low, close=df.Close))
fig.show()
```
